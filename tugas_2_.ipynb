```python
import matplotlib.pyplot as plt
import numpy as np

# Membuat data dummy untuk x dan y
p = df.mean()
q = df.std()

# Membagi data menjadi kelompok-kelompok 10
x_groups = [p[i:i+10] for i in range(0, len(p), 10)]
y_groups = [q[i:i+10] for i in range(0, len(q), 10)]

# Daftar warna yang akan digunakan
colors = ['r', 'g', 'b']

# Membuat scatter plot dengan memberikan warna yang berbeda untuk setiap kelompok 10 data
for i in range(len(x_groups)):
    plt.scatter(x_groups[i], y_groups[i], color=colors[i % len(colors)], label=f'Data {i+1}-{i+10}')

plt.xlabel('Variabel X')
plt.ylabel('Variabel Y')
plt.title('Scatter Plot dengan Warna Berbeda Setiap 10 Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Membuat data dummy untuk x dan y
h = df.skew()
j = df.median()

# Membagi data menjadi kelompok-kelompok 10
x_groups = [h[i:i+10] for i in range(0, len(h), 10)]
y_groups = [j[i:i+10] for i in range(0, len(j), 10)]

# Daftar warna yang akan digunakan
colors = ['r', 'g', 'b']

# Membuat scatter plot dengan memberikan warna yang berbeda untuk setiap kelompok 10 data
for i in range(len(x_groups)):
    plt.scatter(x_groups[i], y_groups[i], color=colors[i % len(colors)], label=f'Data {i+1}-{i+10}')

plt.xlabel('Variabel X')
plt.ylabel('Variabel Y')
plt.title('Scatter Plot dengan Warna Berbeda Setiap 10 Data')
plt.legend()
plt.grid(True)
plt.show()
```
